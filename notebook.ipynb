```python
fullData = s_d.merge(m_d, right_index = True, left_on = '#SampleID')
fullData
```

```python
mask = fullData['SubjectID'] != 'mock'
fullData = fullData[mask]
mask2 = fullData['SubjectID'] != 'NTC'
fullData = fullData[mask2]
```

```python
fullData = fullData[~fullData['SubjectID'].isin(['10015', '10010', '19003', '10011', '10035', '10007', '10012', '10002', '10001'])]
```

```python
fullData = fullData[fullData['GestWeekColl'] < 5000]
```

```python
fullData['History_of_preterm_delivery']= fullData['History_of_preterm_delivery'].fillna(0, inplace = True)
fullData['PPROM']=fullData['PPROM'].fillna(0, inplace = True)
fullData['Chorioamnionitis']=fullData['Chorioamnionitis'].fillna(0, inplace = True)
fullData['Endometritis']=fullData['Endometritis'].fillna(0, inplace = True)
fullData['Hypertensive Disorder']=fullData['Hypertensive Disorder'].fillna(0, inplace = True)
fullData['Preeclampsia']=fullData['Preeclampsia'].fillna(0, inplace = True)
fullData['Number_of_babies']=fullData['Number_of_babies'].fillna(1, inplace = True)
# todo: changes PPROM Yes to 1 and no to 0
def PPROM(row):
    if row['PPROM'] == 'Yes':
        return 1
        #print(1)
    elif row['PPROM'] == 'No':
        return 0
        #print(0)

fullData['PPROM'] = fullData.apply(PPROM, axis = 1)

# todo: change strings in hypertensive disorder where 0 is none/NA, 1 is mild, 2 is severe, 3 is chronic
def hypertensiveD(row):
    if row['Hypertensive Disorder'] == 'None':
        return 0
    elif row['Hypertensive Disorder'] == 'Mild gestational hypertension' or row['Hypertensive Disorder'] == 'Mild Pre-eclampsia':
        return 1
    elif row['Hypertensive Disorder'] == 'Severe Pre-eclampsia':
        return 2
    elif row['Hypertensive Disorder'] == 'Chronic Hypertension': 
        return 3
fullData['Hypertensive Disorder'] = fullData.apply(hypertensiveD, axis = 1)

# todo: change preeclampsia to 1 and 0 instead of T and F
def preeC(row):
    if row['Preeclampsia'] == 'T':
        return 1
    elif row['Preeclampsia'] == 'F':
        return 0
fullData['Preeclampsia'] = fullData.apply(preeC, axis = 1)

# todo: if more than 1 child born, change to 1, if only 1, change to 0
def baby(row):
    if row['Number_of_babies'] is None:
        return 2
    else:
        if row['Number_of_babies'] >= 2:
            return 1
        elif row['Number_of_babies'] == 1:
            return 0
fullData['Binary_baby'] = fullData.apply(baby, axis = 1)

#classify as preterm/not preterm
def preterm(row):
    if row['GestWeekdel'] <= 37:
        return 1
    elif row['GestWeekdel'] > 37:
        return 0
fullData['Preterm'] = fullData.apply(preterm, axis = 1)


fullData.to_csv(basedir + '\Data\processedFullData.csv')
```

```python
#for each sample, this is the number of unique otus
#(m_d > 0).sum(axis = 1)
OTU_cols = fullData.columns[fullData.columns.str.startswith("OTU")]
fullData['Unique_OTU'] = (fullData[OTU_cols] > 0).sum(axis = 1)

mean_OTU = fullData.groupby('SubjectID')['Unique_OTU'].mean()

def average_OTU(row):
    return mean_OTU[row['SubjectID']]
                        
fullData['Mean_OTU'] = fullData.apply(average_OTU,axis=1)
```

```python
#for s, g in fullData.groupby('SubjectID'):
#    print(g['Unique_OTU'].mean())
#    display(g)

import statistics

def standard_dev(row):
    sample = row[OTU_cols]
    dev = statistics.stdev(sample)
    
    return dev

fullData['standard_dev'] = fullData.apply(standard_dev, axis=1)

OTU_count = fullData['Unique_OTU'].sum()

# todo: think about diversity counts
# standard deviation, absolute numbers, skew
# shannon diversity index is total number of species * log of number of species
def shannon_diversity(row):
    diversity = 0
    for col in OTU_cols:
        if row[col]>0:
            p = row[col]/OTU_count
            diversity += p*np.log(p)
    
    diversity = diversity * -1
    return diversity

fullData['shannon_div'] = fullData.apply(shannon_diversity, axis=1)
```

```python
features = fullData[['SubjectID','BodySite','History_of_preterm_delivery',
                     'GestWeekdel','PPROM',
                     'Hypertensive Disorder','Preeclampsia','Number_of_babies',
                     'Mean_OTU','Unique_OTU','shannon_div','standard_dev','Preterm']]

labels = fullData[['SubjectID','BodySite','Preterm']]
```

```python
from scipy import stats
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
X = features[['Mean_OTU','Unique_OTU']]
y = labels['Preterm']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)



from sklearn.neighbors import KNeighborsClassifier

k_range = range(1,5)
scores = {}
scores_list = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train.values.ravel())
    y_pred_knn = knn.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test,y_pred_knn)
    scores_list.append(metrics.accuracy_score(y_test,y_pred_knn))
    
print('Accuracy:',scores_list)
```

```python
# Import necessary package
from sklearn.tree import DecisionTreeClassifier


# This is where we set our machine learning algorithm type - Decision Tree
dectree = DecisionTreeClassifier(criterion='entropy',random_state=0) 
dectree.fit(X_train,y_train.values.ravel())

# Use the model to make predictions using our testing input data
y_pred_dectree = dectree.predict(X_test)

# Calculate the accuracy of the model as a percent
accuracy = metrics.accuracy_score(y_test, y_pred_dectree)
accuracy_percentage = 100 * accuracy
print('Accuracy: ',accuracy_percentage)
```
